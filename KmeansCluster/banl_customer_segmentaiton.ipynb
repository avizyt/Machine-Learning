```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
```

```python
file_path = '../datasets/german.data-numeric'

df = pd.read_csv(file_path,header=None, sep='\s\s+', prefix='X')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df[['X3', 'X9']]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_scaled = scale.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans

kmean = KMeans(random_state=12)

kmean.fit(X_scaled)

preds = kmean.predict(X_scaled)
```

```python
preds
```

```python
df['cluster'] = preds
df.head()
```

```python
df.pop('clusters')
```

```python
df.head()
```

```python
data_pivot = df.pivot_table(
    values=[
        'X3',
        'X9'
    ],
    index='cluster',
    aggfunc=np.mean
    )
```

```python
data_pivot.head()
```

```python
df.loc[df.cluster==0].count()
```

```python
clusters = pd.DataFrame()
inertia = []
```

```python
clusters['cluster_range'] = range(1,15)
```

```python
for k in clusters['cluster_range']:
    kmeans = KMeans(n_clusters=k, random_state=12).fit(X_scaled)
    inertia.append(kmean.inertia_)
```

```python
clusters['inertia'] = inertia
```

```python
clusters
```

```python
clusters
```

```python
alt.Chart(clusters).mark_line().encode(alt.X('cluster_range'), alt.Y('inertia'))
```
