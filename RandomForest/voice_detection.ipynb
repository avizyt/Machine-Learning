```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
random_seed = 1124

X_train, X_test, y_train, y_test = train_test_split(
    data,
    targets, 
    test_size=0.2,
    random_state=random_seed
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rf_model1 = RandomForestClassifier()
```

```python
# fit the model with the train data
rf_model1.fit(X_train, y_train)

# prediction from training data
preds_train = rf_model1.predict(X_train)

# accuracy score from the traing data
acc_train = accuracy_score(y_train, preds_train)

# prediction from test data
preds_test = rf_model1.predict(X_test)

# accuracy score from the test data
acc_test = accuracy_score(y_test, preds_test)

print(f"Train data Accuracy score: {acc_train}")
print(f"Test data Accuracy score: {acc_test}")
print(f"Gap between Train and test accuracy: {acc_train - acc_test}")
```

```python
rf_model2 = RandomForestClassifier(
    n_estimators=20, 
    max_depth=5, 
    min_samples_leaf=10, 
    max_features=0.5
)
```

```python
def modelTrain(model, X_train, y_train, X_test, y_test):
    # fit the model with the train data
    model.fit(X_train, y_train)

    # prediction from training data
    preds_train = model.predict(X_train)

    # accuracy score from the traing data
    acc_train = accuracy_score(y_train, preds_train)

    # prediction from test data
    preds_test = model.predict(X_test)

    # accuracy score from the test data
    acc_test = accuracy_score(y_test, preds_test)

    print(f"Train data Accuracy score: {acc_train}")
    print(f"Test data Accuracy score: {acc_test}")
    print(f"Gap between Train and test accuracy: {acc_train - acc_test}")
```

```python
modelTrain(rf_model2,X_train, y_train, X_test, y_test)
```
